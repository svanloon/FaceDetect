## Facial Reconition

```python
import cv2
import sys
import matplotlib.pyplot as plt
%matplotlib inline

def detect(image, cascPath):
    # Create the haar cascade
    faceCascade = cv2.CascadeClassifier(cascPath)

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
        flags = 0
    )
    return faces


def box(faces, image):
    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

def displayImage(image):
    b,g,r = cv2.split(image)
    img_repacked = cv2.merge([r,g,b])
    plt.imshow(img_repacked) # expect true color
    plt.show()
```

```python
def processFaces(imagePath):
    cascPath = "haarcascades/haarcascade_frontalface_default.xml"
    #cascPath = "haarcascades/haarcascade_frontalcatface_extended.xml"

    # Read the image
    image = cv2.imread(imagePath)
    faces = detect(image, cascPath)
    print "Found {0} faces!".format(len(faces))
    imageWithBoxes = image.copy()
    box(faces, imageWithBoxes)
    displayImage(imageWithBoxes)
    displayImage(image)
```

```python
processFaces("abba.png")
processFaces("DSC_1528.JPG")
```
